```python
selected_fea=['genres','keywords','tagline','cast','director','overview']
print(selected_fea)
```

```python
#Combining all the selected features
combine_fea=df['genres']+df['keywords']+df['tagline']+df['cast']+df['director']+df['overview']
print(combine_fea)
```

```python
#Converting the text data to feature vectors
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
fea_vector=vectorizer.fit_transform(combine_fea)
print(fea_vector)
```

# Cosine similarity

```python
#Getting the similarity scores using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(fea_vector) #checking similarity 
print(similarity)
```

```python
print(similarity.shape) #each movie compare with all movies in datasets
```

# Getting movie name from user

```python
#Getting the movie name from the user
movie_name=input('Enter your favourite movie name:')
```

```python
#creating a list with all the movie names given in the dataset
list_of_all_titles=df['title'].tolist()
print(list_of_all_titles)
```

```python
#Finding the close match for the movie name givenby the user
find_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_match)
```

```python
close_match=find_match[0]
print(close_match)
```

```python
#Finding the index of the movie with title
index_movie=df[df.title==close_match]['index'].values[0]
print(index_movie)
```

```python
#Getting a list of similar movies
similarity_score=list(enumerate(similarity[index_movie])) #comparing the value of all the given movie[index and similarity score]
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#Sorting the movies based on their similarity score
sorted_similar=sorted(similarity_score,key=lambda x:x[1],reverse=True) #descending order score
print(sorted_similar)
```

```python
#Display the name of similar movies based on the index
print('Movies suggested for you: ')

i=1
for movie in sorted_similar:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0] #check the title of the movie
    if(i<=15):# first 15 movies
        print(i,'-',title_from_index)
        i+=1
    
```

# Movie Recommendation system


```python
movie_name=input('Enter your favourite movie name :')
list_of_all_titles=df['title'].tolist()
find_match=difflib.get_close_matches(movie_name,list_of_all_titles)
close_match=find_match[0]
index_movie=df[df.title==close_match]['index'].values[0]
similarity_score=list(enumerate(similarity[index_movie]))
sorted_similar=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print('Movies suggested for you.. ')

i=1
for movie in sorted_similar:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0] #check the title of the movie
    if(i<=15):# first 15 movies
        print(i,'-',title_from_index)
        i+=1
    
```
